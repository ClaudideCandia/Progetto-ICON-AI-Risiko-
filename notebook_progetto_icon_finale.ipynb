```python
%%writefile risiko_obbiettivi.pl

% ==============================================================================
% SEZIONE 0: FATTI DINAMICI
% ==============================================================================
% Questi predicati verranno aggiunti/rimossi da Python durante il gioco.
% Formato: occupato(Territorio, Giocatore, NumeroArmate).
:- dynamic occupato/3.

% ==============================================================================
% SEZIONE 4: OBIETTIVI E VERIFICA
% ==============================================================================

% --- PREDICATI AUSILIARI (Helpers) ---\

% Conta quanti territori di un Continente NON sono posseduti dal Giocatore.
territori_mancanti_in_continente(Giocatore, Continente, NumMancanti) :-
    % Trova tutti i territori del continente che NON sono occupati dal Giocatore
    findall(T, (territorio(T, Continente), \+ occupato(T, Giocatore, _)), ListaMancanti),
    length(ListaMancanti, NumMancanti).

% Verifica se un giocatore possiede interamente un continente (0 mancanti).
possiede_continente(Giocatore, Continente) :-
    territori_mancanti_in_continente(Giocatore, Continente, 0).

% Trova il continente opzionale più facile da conquistare
miglior_terzo_continente(Giocatore, C1, C2, MigliorC, MinMancanti) :-
    findall(M-C, (
        territorio(_, C),           % Genera un continente C
        C \= C1, C \= C2,           % Assicurati che non sia uno dei due fissi
        territori_mancanti_in_continente(Giocatore, C, M) % Calcola quanto manca
    ), ListaCoppie),
    keysort(ListaCoppie, [MinMancanti-MigliorC | _]). % Ordina per mancanti crescenti e prendi il primo


% --- DEFINIZIONE OBIETTIVI ---\

% 1. OBIETTIVO: ELIMINA COLORE
obiettivo_raggiunto(_, distruggi_giocatore(Target)) :-
    \+ occupato(_, Target, _).

quanto_manca(_, distruggi_giocatore(Target), NumTerritoriRimasti) :-
    findall(T, occupato(T, Target, _), ListaTerritori),
    length(ListaTerritori, NumTerritoriRimasti).

% 2. OBIETTIVO: CONQUISTA 2 CONTINENTI SPECIFICI
obiettivo_raggiunto(Giocatore, conquista_continenti(C1, C2)) :-
    possiede_continente(Giocatore, C1),
    possiede_continente(Giocatore, C2).

quanto_manca(Giocatore, conquista_continenti(C1, C2), TotaleMancanti) :-
    territori_mancanti_in_continente(Giocatore, C1, M1),
    territori_mancanti_in_continente(Giocatore, C2, M2),
    TotaleMancanti is M1 + M2.

% 3. OBIETTIVO: CONQUISTA 2 CONTINENTI + 1 A SCELTA
obiettivo_raggiunto(Giocatore, conquista_continenti(C1, C2, terzo_continente)) :-
    possiede_continente(Giocatore, C1),
    possiede_continente(Giocatore, C2),
    territorio(_, C3),
    C3 \= C1, C3 \= C2,
    possiede_continente(Giocatore, C3).

quanto_manca(Giocatore, conquista_continenti(C1, C2, terzo_continente), TotaleMancanti) :-
    territori_mancanti_in_continente(Giocatore, C1, M1),
    territori_mancanti_in_continente(Giocatore, C2, M2),
    miglior_terzo_continente(Giocatore, C1, C2, _, M3),
    TotaleMancanti is M1 + M2 + M3.

% ### NUOVO: 4. OBIETTIVO NUMERICO (Es. 24 Territori) ###
obiettivo_raggiunto(Giocatore, conquista_territori(NumeroTarget)) :-
    findall(T, occupato(T, Giocatore, _), IMieiTerritori),
    length(IMieiTerritori, Posseduti),
    Posseduti >= NumeroTarget.

quanto_manca(Giocatore, conquista_territori(NumeroTarget), Mancanti) :-
    findall(T, occupato(T, Giocatore, _), IMieiTerritori),
    length(IMieiTerritori, Posseduti),
    (Posseduti >= NumeroTarget -> Mancanti = 0 ; Mancanti is NumeroTarget - Posseduti).

% ### NUOVO: 5. OBIETTIVO PRESIDIO (Es. 18 Territori con 2 Armate) ###
obiettivo_raggiunto(Giocatore, conquista_territori_armate(NumTarget, MinArmate)) :-
    findall(T, (
        occupato(T, Giocatore, ArmatePresenti),
        ArmatePresenti >= MinArmate
    ), TerritoriValidi),
    length(TerritoriValidi, Conteggio),
    Conteggio >= NumTarget.

quanto_manca(Giocatore, conquista_territori_armate(NumTarget, MinArmate), Mancanti) :-
    findall(T, (
        occupato(T, Giocatore, ArmatePresenti),
        ArmatePresenti >= MinArmate
    ), TerritoriValidi),
    length(TerritoriValidi, Conteggio),
    (Conteggio >= NumTarget -> Mancanti = 0 ; Mancanti is NumTarget - Conteggio).


% --- IDENTIFICAZIONE TERRITORI DELL'OBIETTIVO (SCOPE) ---\

% CASO A: Obiettivo Distruggi Giocatore
territorio_obiettivo(_, distruggi_giocatore(Nemico), Territorio) :-
    occupato(Territorio, Nemico, _).

% CASO B: Obiettivo Continenti
territorio_obiettivo(_, conquista_continenti(C1, C2), Territorio) :-
    (territorio(Territorio, C1) ; territorio(Territorio, C2)).

% CASO C: Obiettivo Conquista 2 Continenti + 1 A Scelta
territorio_obiettivo(Giocatore, conquista_continenti(C1, C2, terzo_continente), Territorio) :-
    (territorio(Territorio, C1) ; territorio(Territorio, C2)).

territorio_obiettivo(Giocatore, conquista_continenti(C1, C2, terzo_continente), Territorio) :-
    miglior_terzo_continente(Giocatore, C1, C2, MigliorC, _),
    territorio(Territorio, MigliorC).

% ### NUOVO: CASO D - Obiettivo Numerico (Tutto il mondo non mio è target) ###
territorio_obiettivo(Giocatore, conquista_territori(_), Territorio) :-
    \+ occupato(Territorio, Giocatore, _).

% ### NUOVO: CASO E - Obiettivo Presidio ###
% Sono obiettivi sia i territori nemici (da prendere)
% sia i MIEI territori che però hanno meno armate del richiesto (da rinforzare).
territorio_obiettivo(Giocatore, conquista_territori_armate(_, MinArmate), Territorio) :-
    \+ occupato(Territorio, Giocatore, _). % Territorio Nemico

territorio_obiettivo(Giocatore, conquista_territori_armate(_, MinArmate), Territorio) :-
    occupato(Territorio, Giocatore, Armate),
    Armate < MinArmate. % Territorio Mio ma debole
```

```python
%%writefile risiko_tattica.pl

% ==============================================================================
% SEZIONE TATTICA: FRONTE, RETROVIA, VULNERABILITÀ
% ==============================================================================

% Dichiariamo i fatti dinamici che verranno popolati da Python
:- dynamic occupato/3.
:- dynamic conquista_effettuata/1. % True se il giocatore ha preso un territorio nel turno
:- dynamic bonus_carte_attivo/1.   % True se il giocatore ha un Tris pronto da giocare
:- dynamic carte_giocatore/2.      % Definito in analisi_globale, ma utile dichiararlo qui per sicurezza

% --- HELPER: Analisi dei vicini ---
max_minaccia_vicina(Territorio, Giocatore, MaxArmateNemico) :-
    findall(Armate, (
        adiacente(Territorio, NemicoPos),
        occupato(NemicoPos, Avversario, Armate),
        Avversario \= Giocatore
    ), ListaArmateNemiche),
    (ListaArmateNemiche = [] -> MaxArmateNemico = 0 ; max_list(ListaArmateNemiche, MaxArmateNemico)).

% --- 1. IL FRONTE (Prima Linea) ---
fronte(Territorio, Giocatore) :-
    occupato(Territorio, Giocatore, _),
    adiacente(Territorio, Vicino),
    occupato(Vicino, Nemico, _),
    Giocatore \= Nemico.

% --- 2. LA RETROVIA (Seconda Linea) ---
retrovia(Territorio, Giocatore) :-
    occupato(Territorio, Giocatore, _),
    \+ fronte(Territorio, Giocatore).

% --- 3. TERRITORIO VULNERABILE ---
vulnerabile(Territorio, Giocatore) :-
    fronte(Territorio, Giocatore),
    occupato(Territorio, Giocatore, MieArmate),
    max_minaccia_vicina(Territorio, Giocatore, MaxNem),
    MaxNem > 0,
    Soglia is MaxNem / 2,
    MieArmate < Soglia.

% --- 4. TERRITORIO FORTE AL FRONTE ---
forte_al_fronte(Territorio, Giocatore) :-
    fronte(Territorio, Giocatore),
    occupato(Territorio, Giocatore, MieArmate),
    max_minaccia_vicina(Territorio, Giocatore, MaxNem),
    Soglia is MaxNem * 2,
    MieArmate > Soglia.

% --- 5. LOGICA "CARD FARMING" (Forza Attacco) ---
% L'agente deve forzare un attacco se:
% 1. NON ha ancora conquistato nulla in questo turno.
% 2. HA almeno un territorio forte al fronte (quindi ha risorse da spendere).
% 3. HA bisogno di carte (ne ha esattamente 2 OPPURE non ha un bonus attivo).
forza_attacco(Giocatore) :-
    \+ conquista_effettuata(Giocatore),
    forte_al_fronte(_, Giocatore), % Esiste almeno un territorio forte
    (carte_giocatore(Giocatore, 2) ; \+ bonus_carte_attivo(Giocatore)).

% ==============================================================================
% SEZIONE 6: GESTIONE CARTE (Tris)
% ==============================================================================

% Decide se è il momento giusto per giocare il tris.
% Non lo gioca se è in modalità 'consolidamento' e ha poche carte (aspetta un tris migliore),
% a meno che non sia costretto.

conviene_giocare_tris(Giocatore) :-
    bonus_carte_attivo(Giocatore), % Condizione necessaria: avere il tris
    (
        % A. CASO OBBLIGATO/EFFICIENZA: Ho 5 o più carte (rischio stallo o regola)
        (carte_giocatore(Giocatore, N), N >= 5);

        % B. CASO EMERGENZA: Rischio di morire, mi serve tutto l'aiuto possibile
        rischio_eliminazione(Giocatore);

        % C. CASO AGGRESSIVO: Voglio attaccare, mi servono munizioni
        modalita_attiva(Giocatore, dominio);

        % D. CASO STANDARD: Espansione (più truppe = più territori)
        modalita_attiva(Giocatore, espansione);

        % E. CASO RECUPERO: Sono indietro, devo colmare il gap
        in_svantaggio(Giocatore)
    ).

% Nota: Se sono in 'consolidamento' e ho < 5 carte, la regola fallisce (risparmio il tris).
```

```python
%%writefile risiko_analisi_globale.pl

% ==============================================================================
% SEZIONE 5: ANALISI GLOBALE E STATO GIOCATORI
% ==============================================================================
% Dipendenze:
% - occupato/3 (Dinamico)
% - carte_giocatore/2 (Dinamico)
% - intercontinentale/1 (da risiko_geografia.pl)
% - territorio/2 (da risiko_geografia.pl)
% - vulnerabile/2 (da risiko_tattica.pl)

:- dynamic occupato/3.
:- dynamic carte_giocatore/2.

% --- CALCOLO DEL FATTORE MOLTIPLICATIVO CARTE ---
fattore_carte(NumCarte, 1.0) :- NumCarte =< 2.
fattore_carte(NumCarte, 2.0) :- NumCarte >= 3, NumCarte =< 4.
fattore_carte(NumCarte, 3.0) :- NumCarte >= 5.

% --- CALCOLO DEL PUNTEGGIO DI POTENZA ---
calcola_punteggio(Giocatore, ScoreFinale) :-
    occupato(_, Giocatore, _),

    % 1. Calcolo Score Base
    findall(T, (occupato(T, Giocatore, _), \+ intercontinentale(T)), ListaNorm), length(ListaNorm, NumNorm),
    findall(Tp, (occupato(Tp, Giocatore, _), intercontinentale(Tp)), ListaPonte), length(ListaPonte, NumPonte),
    findall(A, occupato(_, Giocatore, A), ListaArmate), sum_list(ListaArmate, TotArmate),
    % Qui chiamiamo la regola che prima falliva
    findall(C, possiede_intero_continente_check(Giocatore, C), ListaCont), length(ListaCont, NumCont),

    BaseScore is (NumNorm * 1.0) + (NumPonte * 1.5) + (TotArmate * 1.5) + (NumCont * 2.0),

    % 2. Recupero Carte e Applicazione Moltiplicatore
    (carte_giocatore(Giocatore, N) -> NumCarte = N ; NumCarte = 0),
    fattore_carte(NumCarte, Moltiplicatore),

    ScoreFinale is BaseScore * Moltiplicatore.

% --- Helper CORRETTO per verificare possesso continente ---
possiede_intero_continente_check(Giocatore, Continente) :-
    continente(Continente, _),
    \+ (territorio(T, Continente), \+ occupato(T, Giocatore, _)).


% --- DEFINIZIONE DEGLI STATI GIOCATORE ---

dominante(Giocatore) :-
    calcola_punteggio(Giocatore, MioScore),
    forall((occupato(_, Altro, _), Altro \= Giocatore),
           (calcola_punteggio(Altro, AltroScore), MioScore > AltroScore)).

in_svantaggio(Giocatore) :-
    calcola_punteggio(Giocatore, MioScore),
    forall((occupato(_, Altro, _), Altro \= Giocatore),
           (calcola_punteggio(Altro, AltroScore), MioScore < AltroScore)).

normale(Giocatore) :-
    occupato(_, Giocatore, _),
    \+ dominante(Giocatore),
    \+ in_svantaggio(Giocatore).

rischio_eliminazione(Giocatore) :-
    occupato(_, Giocatore, _),
    findall(T, occupato(T, Giocatore, _), Territori),
    length(Territori, N),
    N < 3.


% --- ANALISI STATO CONTINENTI ---

continente_conteso(Continente) :-
    continente(Continente, _),
    setof(Giocatore, T^(territorio(T, Continente), occupato(T, Giocatore, _)), ListaProprietari),
    length(ListaProprietari, N),
    N > 1.

continente_fortezza(Continente, Giocatore) :-
    possiede_intero_continente_check(Giocatore, Continente),
    forall(
        (territorio(T, Continente), intercontinentale(T)),
        \+ vulnerabile(T, Giocatore)
    ).
```

```python
%%writefile risiko_comportamento.pl

% ==============================================================================
% SEZIONE 6: COMPORTAMENTO E PRIORITÀ STRATEGICHE
% ==============================================================================
% Dipendenze logiche (devono essere caricate le altre KB):
% - rischio_eliminazione/1, in_svantaggio/1, dominante/1 (da analisi_globale)
% - vulnerabile/2, forte_al_fronte/2 (da tattica)

% --- 1. DETERMINAZIONE DELLA MODALITÀ DI GIOCO ---
% Definisce "l'umore" dell'agente in base allo stato globale.

modalita_attiva(Giocatore, sopravvivenza) :-
    rischio_eliminazione(Giocatore).

modalita_attiva(Giocatore, consolidamento) :-
    in_svantaggio(Giocatore),
    \+ rischio_eliminazione(Giocatore).

modalita_attiva(Giocatore, dominio) :-
    dominante(Giocatore).

% Default: Se non è in nessuno degli stati sopra, è in espansione
modalita_attiva(Giocatore, espansione) :-
    normale(Giocatore).


% --- 2. ASSEGNAZIONE PESI ALLE AZIONI (Policy) ---
% peso_azione(Modalita, TipoAzione, Peso).
% I pesi vanno da 0 a 100.

% --- A. MODALITÀ SOPRAVVIVENZA (Tartaruga) ---
% Obiettivo: Non morire. Difendere tutto. Non attaccare a meno che sia gratis.
peso_azione(sopravvivenza, rinforzo_difensivo, 100).   % Massima priorità
peso_azione(sopravvivenza, rinforzo_offensivo, 10).    % Inutile ora
peso_azione(sopravvivenza, attacco_sicuro, 20).        % Solo se garantito
peso_azione(sopravvivenza, attacco_rischioso, 0).      % MAI
peso_azione(sopravvivenza, spostamento_fronte, 80).    % Portare tutto al fronte

% --- B. MODALITÀ CONSOLIDAMENTO (Costruttore) ---
% Obiettivo: Uscire dallo svantaggio prendendo continenti piccoli.
peso_azione(consolidamento, rinforzo_difensivo, 80).
peso_azione(consolidamento, rinforzo_offensivo, 40).
peso_azione(consolidamento, attacco_sicuro, 60).
peso_azione(consolidamento, attacco_rischioso, 10).
peso_azione(consolidamento, spostamento_fronte, 60).

% --- C. MODALITÀ ESPANSIONE (Equilibrato) ---
% Obiettivo: Gioco standard. Bilanciamento tra obiettivi e difesa.
peso_azione(espansione, rinforzo_difensivo, 50).
peso_azione(espansione, rinforzo_offensivo, 70).       % Preparare attacchi è importante
peso_azione(espansione, attacco_sicuro, 80).           % Prendere territori facili
peso_azione(espansione, attacco_rischioso, 30).        % Si può osare ogni tanto
peso_azione(espansione, spostamento_fronte, 50).

% --- D. MODALITÀ DOMINIO (Schiacciasassi) ---
% Obiettivo: Annientare. Spingere tutto sull'attacco.
peso_azione(dominio, rinforzo_difensivo, 20).          % La miglior difesa è l'attacco
peso_azione(dominio, rinforzo_offensivo, 100).         % Caricare i cannoni
peso_azione(dominio, attacco_sicuro, 100).             % Mangiare tutto
peso_azione(dominio, attacco_rischioso, 60).           % Osare per chiudere
peso_azione(dominio, spostamento_fronte, 40).


% --- 3. REGOLA INTERFACCIA (API) ---
% Questa è la regola che Python interrogherà.
% Restituisce la priorità di una specifica azione per il giocatore corrente.

ottieni_priorita(Giocatore, TipoAzione, Peso) :-
    modalita_attiva(Giocatore, Modalita),
    peso_azione(Modalita, TipoAzione, Peso).

% Debug: Per capire in che modalità è l'agente
debug_modalita(Giocatore, Modalita) :-
    modalita_attiva(Giocatore, Modalita).
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import time
from collections import Counter

# =============================================================================
# 1. RISK BOARD (Invariato - riporto per completezza di esecuzione)
# =============================================================================

_NO_UPDATE = object()

class RiskBoard:
    def __init__(self):
        self.graph = nx.Graph()
        self._initialize_map()

    def _initialize_map(self):
        self.continents = {
            'nord_america': 5, 'sud_america': 2, 'europa': 5,
            'africa': 3, 'asia': 7, 'oceania': 2
        }
        territories_data = [
            ("alaska", "nord_america", ["kamchatka", "alberta", "nord_ovest"]),
            ("nord_ovest", "nord_america", ["alaska", "alberta", "ontario", "groenlandia"]),
            ("groenlandia", "nord_america", ["nord_ovest", "ontario", "quebec", "islanda"]),
            ("alberta", "nord_america", ["alaska", "nord_ovest", "ontario", "stati_uniti_occidentali"]),
            ("ontario", "nord_america", ["alberta", "nord_ovest", "groenlandia", "quebec", "stati_uniti_occidentali", "stati_uniti_orientali"]),
            ("quebec", "nord_america", ["ontario", "groenlandia", "stati_uniti_orientali"]),
            ("stati_uniti_occidentali", "nord_america", ["alberta", "ontario", "stati_uniti_orientali", "america_centrale"]),
            ("stati_uniti_orientali", "nord_america", ["ontario", "quebec", "stati_uniti_occidentali", "america_centrale"]),
            ("america_centrale", "nord_america", ["stati_uniti_occidentali", "stati_uniti_orientali", "venezuela"]),
            ("venezuela", "sud_america", ["america_centrale", "peru", "brasile"]),
            ("peru", "sud_america", ["venezuela", "brasile", "argentina"]),
            ("brasile", "sud_america", ["venezuela", "peru", "argentina", "nord_africa"]),
            ("argentina", "sud_america", ["peru", "brasile"]),
            ("islanda", "europa", ["groenlandia", "gran_bretagna", "scandinavia"]),
            ("scandinavia", "europa", ["islanda", "gran_bretagna", "europa_settentrionale", "ucraina"]),
            ("gran_bretagna", "europa", ["islanda", "scandinavia", "europa_settentrionale", "europa_occidentale"]),
            ("europa_settentrionale", "europa", ["scandinavia", "gran_bretagna", "europa_occidentale", "europa_meridionale", "ucraina"]),
            ("europa_occidentale", "europa", ["gran_bretagna", "europa_settentrionale", "europa_meridionale", "nord_africa"]),
            ("europa_meridionale", "europa", ["europa_occidentale", "europa_settentrionale", "ucraina", "nord_africa", "egitto", "medio_oriente"]),
            ("ucraina", "europa", ["scandinavia", "europa_settentrionale", "europa_meridionale", "medio_oriente", "afghanistan", "urali"]),
            ("nord_africa", "africa", ["brasile", "europa_occidentale", "europa_meridionale", "egitto", "africa_orientale", "congo"]),
            ("egitto", "africa", ["europa_meridionale", "nord_africa", "africa_orientale", "medio_oriente"]),
            ("africa_orientale", "africa", ["egitto", "nord_africa", "congo", "sud_africa", "madagascar"]),
            ("congo", "africa", ["nord_africa", "africa_orientale", "sud_africa"]),
            ("sud_africa", "africa", ["congo", "africa_orientale", "madagascar"]),
            ("madagascar", "africa", ["africa_orientale", "sud_africa"]),
            ("urali", "asia", ["ucraina", "afghanistan", "siberia", "cina"]),
            ("siberia", "asia", ["urali", "jacuzia", "cita", "mongolia", "cina"]),
            ("jacuzia", "asia", ["siberia", "cita", "kamchatka"]),
            ("kamchatka", "asia", ["jacuzia", "cita", "mongolia", "giappone", "alaska"]),
            ("cita", "asia", ["siberia", "jacuzia", "kamchatka", "mongolia", "giappone"]),
            ("mongolia", "asia", ["siberia", "cita", "kamchatka", "giappone", "cina"]),
            ("giappone", "asia", ["kamchatka", "cita", "mongolia"]),
            ("afghanistan", "asia", ["ucraina", "urali", "cina", "india", "medio_oriente"]),
            ("cina", "asia", ["mongolia", "siberia", "urali", "afghanistan", "india", "siam"]),
            ("india", "asia", ["medio_oriente", "afghanistan", "cina", "siam"]),
            ("siam", "asia", ["cina", "india", "indonesia"]),
            ("medio_oriente", "asia", ["europa_meridionale", "ucraina", "egitto", "india", "afghanistan", "cina"]),
            ("indonesia", "oceania", ["siam", "nuova_guinea", "australia_occidentale"]),
            ("nuova_guinea", "oceania", ["indonesia", "australia_occidentale", "australia_orientale"]),
            ("australia_occidentale", "oceania", ["indonesia", "nuova_guinea", "australia_orientale"]),
            ("australia_orientale", "oceania", ["nuova_guinea", "australia_occidentale"])
        ]
        for name, cont, neighbors in territories_data:
            self.graph.add_node(name, continent=cont, owner=None, armies=0)
            for neighbor in neighbors:
                if not self.graph.has_edge(name, neighbor): self.graph.add_edge(name, neighbor)

    def get_territory(self, name): return self.graph.nodes[name]
    def get_owner(self, name): return self.graph.nodes[name]['owner']
    def get_armies(self, name): return self.graph.nodes[name]['armies']
    def get_neighbors(self, name): return list(self.graph.neighbors(name))
    def get_territories_by_player(self, player): return [n for n, d in self.graph.nodes(data=True) if d['owner'] == player]
    def get_continent_territories(self, continent_name): return [n for n, attr in self.graph.nodes(data=True) if attr['continent'] == continent_name]
    def is_continent_owned(self, continent_name, player):
        territories = self.get_continent_territories(continent_name)
        if not territories: return False
        return all(self.graph.nodes[t]['owner'] == player for t in territories)
    def check_path(self, source, target, player):
        if self.get_owner(source) != player or self.get_owner(target) != player: return False
        player_nodes = self.get_territories_by_player(player)
        subgraph = self.graph.subgraph(player_nodes)
        return nx.has_path(subgraph, source, target)
    def update_territory(self, name, player=_NO_UPDATE, armies=_NO_UPDATE):
        if player is not _NO_UPDATE: self.graph.nodes[name]['owner'] = player
        if armies is not _NO_UPDATE: self.graph.nodes[name]['armies'] = armies

# =============================================================================
# 2. RISK GAME (CON LOGICA SCAMBIO TRIS)
# =============================================================================

class RiskGame:
    def __init__(self, board=None, players=None):
        self.board = board if board else RiskBoard()
        self.players = players if players else ["blu", "rosso", "verde"]
        self.current_player_idx = 0
        self.turn_phase = 'placement'
        self.troops_to_place = 0
        self.hands = {p: [] for p in self.players}
        self.conquest_this_turn = False
        self.deck = self._initialize_deck()
        if self.board.get_owner("alaska") is None: self._random_populate()
        self._start_turn_reinforcements()

    def _initialize_deck(self):
        deck = []
        deck.extend(["cannone"] * 14)
        deck.extend(["fante"] * 14)
        deck.extend(["cavalleria"] * 14)
        deck.extend(["jolly"] * 2)
        random.shuffle(deck)
        return deck

    def draw_card(self, player):
        if not self.deck: return None
        card = self.deck.pop()
        self.hands[player].append(card)
        return card

    # --- LOGICA TRIS ---

    def has_tris_bonus(self, player):
        """Controlla se esiste una combinazione valida (booleano)."""
        cards = self.hands[player]
        if len(cards) < 3: return False
        counts = Counter(cards)
        n_jolly = counts.get("jolly", 0)
        n_cannone = counts.get("cannone", 0)
        n_fante = counts.get("fante", 0)
        n_cavalleria = counts.get("cavalleria", 0)

        if n_cannone >= 3 or n_fante >= 3 or n_cavalleria >= 3: return True
        if n_cannone >= 1 and n_fante >= 1 and n_cavalleria >= 1: return True
        if n_jolly >= 1 and (n_cannone >= 2 or n_fante >= 2 or n_cavalleria >= 2): return True
        return False

    def redeem_cards(self, player):
        """
        Scambia le carte del giocatore con armate bonus.
        Trova la combinazione MIGLIORE e la rimuove.
        Restituisce le armate guadagnate.
        """
        cards = self.hands[player]
        if len(cards) < 3: return 0

        counts = Counter(cards)
        to_remove = []
        bonus = 0

        # Priorità 1: Jolly + 2 Uguali (12 Armate)
        if counts["jolly"] >= 1:
            for c_type in ["cannone", "fante", "cavalleria"]:
                if counts[c_type] >= 2:
                    to_remove = ["jolly", c_type, c_type]
                    bonus = 12
                    break

        # Priorità 2: 3 Diversi (10 Armate)
        if not to_remove:
            if counts["cannone"] >= 1 and counts["fante"] >= 1 and counts["cavalleria"] >= 1:
                to_remove = ["cannone", "fante", "cavalleria"]
                bonus = 10

        # Priorità 3: 3 Cavalleria (8 Armate)
        if not to_remove and counts["cavalleria"] >= 3:
            to_remove = ["cavalleria"] * 3
            bonus = 8

        # Priorità 4: 3 Fanti (6 Armate)
        if not to_remove and counts["fante"] >= 3:
            to_remove = ["fante"] * 3
            bonus = 6

        # Priorità 5: 3 Cannoni (4 Armate)
        if not to_remove and counts["cannone"] >= 3:
            to_remove = ["cannone"] * 3
            bonus = 4

        # Esecuzione Scambio
        if to_remove:
            # print(f"  [CARTE] {player} scambia il tris {to_remove} per {bonus} armate!")
            for card in to_remove:
                self.hands[player].remove(card)
            self.troops_to_place += bonus
            return bonus

        return 0

    # --- FINE LOGICA TRIS ---

    @property
    def current_player(self): return self.players[self.current_player_idx]

    def _random_populate(self):
        nodes = list(self.board.graph.nodes())
        random.shuffle(nodes)
        for i, node in enumerate(nodes):
            owner = self.players[i % len(self.players)]
            self.board.update_territory(node, player=owner, armies=random.randint(2, 5))

    def _start_turn_reinforcements(self):
        player = self.current_player
        owned = self.board.get_territories_by_player(player)
        base_troops = max(3, len(owned) // 3)
        continent_bonus = 0
        for cont, bonus in self.board.continents.items():
            if self.board.is_continent_owned(cont, player): continent_bonus += bonus
        self.troops_to_place = base_troops + continent_bonus
        self.turn_phase = 'placement'
        self.conquest_this_turn = False

    def place_troops(self, territory, amount):
        if self.turn_phase != 'placement': return False
        if self.board.get_owner(territory) != self.current_player: return False
        if amount > self.troops_to_place: amount = self.troops_to_place
        current = self.board.get_armies(territory)
        self.board.update_territory(territory, armies=current + amount)
        self.troops_to_place -= amount
        if self.troops_to_place == 0: self.turn_phase = 'attack'
        return True

    def validate_attack(self, from_terr, to_terr):
        if self.turn_phase != 'attack': return False, "Fase errata."
        if self.board.get_owner(from_terr) != self.current_player: return False, "Non tuo."
        if self.board.get_owner(to_terr) == self.current_player: return False, "Attacchi te stesso."
        if to_terr not in self.board.get_neighbors(from_terr): return False, "Non adiacente."
        if self.board.get_armies(from_terr) < 2: return False, "Armate insufficienti (serve > 1)."
        return True, "Valid"

    def resolve_battle(self, from_terr, to_terr, verbose=True):
        is_valid, msg = self.validate_attack(from_terr, to_terr)
        if not is_valid: return False
        att_armies = self.board.get_armies(from_terr)
        def_armies = self.board.get_armies(to_terr)
        n_att = min(3, att_armies - 1)
        n_def = min(3, def_armies)
        dadi_att = sorted([random.randint(1, 6) for _ in range(n_att)], reverse=True)
        dadi_def = sorted([random.randint(1, 6) for _ in range(n_def)], reverse=True)
        att_losses, def_losses = 0, 0
        for i in range(min(n_att, n_def)):
            if dadi_def[i] >= dadi_att[i]: att_losses += 1
            else: def_losses += 1
        if verbose: print(f"        🎲 [Dadi] Att:{dadi_att} vs Def:{dadi_def} -> Persi: Att-{att_losses} / Def-{def_losses}")
        new_att_armies = att_armies - att_losses
        new_def_armies = def_armies - def_losses
        self.board.update_territory(from_terr, armies=new_att_armies)
        self.board.update_territory(to_terr, armies=new_def_armies)
        if new_def_armies <= 0:
            if verbose: print(f"        🚩 {to_terr} CONQUISTATO da {self.current_player}!")
            self.conquest_this_turn = True
            move_in = n_att
            self.board.update_territory(from_terr, armies=new_att_armies - move_in)
            self.board.update_territory(to_terr, player=self.current_player, armies=move_in)
            return True
        return False

    def end_attack_phase(self):
        if self.turn_phase == 'attack': self.turn_phase = 'maneuver'

    def validate_fortify(self, from_terr, to_terr):
        if self.turn_phase != 'maneuver': return False
        if self.board.get_owner(from_terr) != self.current_player: return False
        if self.board.get_owner(to_terr) != self.current_player: return False
        if self.board.get_armies(from_terr) <= 1: return False
        return self.board.check_path(from_terr, to_terr, self.current_player)

    def fortify(self, from_terr, to_terr, amount):
        if self.validate_fortify(from_terr, to_terr):
            self.board.update_territory(from_terr, armies=self.board.get_armies(from_terr) - amount)
            self.board.update_territory(to_terr, armies=self.board.get_armies(to_terr) + amount)
            self.end_turn()

    def end_turn(self):
        if self.conquest_this_turn: self.draw_card(self.current_player)
        self.current_player_idx = (self.current_player_idx + 1) % len(self.players)
        self._start_turn_reinforcements()

# =============================================================================
# 3. VISUALIZATION FUNCTIONS (AGGIUNTE)
# =============================================================================

# --- AGGIUNTA: Implementazione visualize_game_state (Colori per Giocatore) ---
def visualize_game_state(risk_board):
    """
    Visualizza il grafo colorando i nodi in base al proprietario (blu, rosso, verde).
    Mostra anche il numero di armate.
    """
    plt.figure(figsize=(15, 10))
    G = risk_board.graph

    # Mappa colori giocatori
    player_colors_map = {
        'blu': '#4169E1',    # RoyalBlue
        'rosso': '#DC143C',  # Crimson
        'verde': '#228B22',  # ForestGreen
        None: '#D3D3D3'      # LightGray (neutrale/errore)
    }

    # Crea lista colori per ogni nodo
    node_colors = [player_colors_map.get(G.nodes[n]['owner']) for n in G.nodes()]

    # Etichette con nome e armate
    labels = {n: f"{n}\n[{G.nodes[n]['armies']}]" for n in G.nodes()}

    # Layout e disegno
    pos = nx.spring_layout(G, k=0.15, seed=42)
    nx.draw_networkx_nodes(G, pos, node_size=1500, node_color=node_colors, edgecolors='black')
    nx.draw_networkx_edges(G, pos, width=1.7, alpha=0.9, edge_color='gray')
    nx.draw_networkx_labels(G, pos, labels, font_size=8, font_color='white', font_weight='bold')

    plt.title("Stato del Gioco - Proprietari & Armate")
    plt.axis('off')
    plt.show()

# --- AGGIUNTA: Implementazione visualize_board (Colori per Continente) ---
def visualize_board(risk_board):
    """
    Visualizza il grafo colorando i nodi in base al continente di appartenenza.
    Schema colori: Nord America (Arancione), Sud America (Celeste),
    Africa (Terra di Siena), Asia (Verde Salvia), Europa (Viola), Oceania (Rosa).
    """
    plt.figure(figsize=(25, 12))
    G = risk_board.graph

    # Mappa colori continenti richiesta
    continent_palette = {
        'nord_america': '#FFA500',   # Arancione
        'sud_america': '#87CEEB',    # Celeste (SkyBlue)
        'africa': '#A0522D',         # Terra di Siena (Sienna)
        'asia': '#8FBC8F',           # Verde Salvia (DarkSeaGreen)
        'europa': '#9370DB',         # Viola (MediumPurple)
        'oceania': '#FF69B4'         # Rosa (HotPink)
    }

    # Crea lista colori nodi basata sull'attributo 'continent'
    # Usa Grigio chiaro come fallback se il continente non è in lista
    node_colors = [continent_palette.get(G.nodes[n]['continent'], '#D3D3D3') for n in G.nodes()]

    # Layout e disegno (uso lo stesso seed per coerenza visiva con l'altra vista)
    pos = nx.spring_layout(G, k=0.8, iterations=300, seed=42)
    nx.draw_networkx_nodes(G, pos, node_size=2500, node_color=node_colors, edgecolors='black')
    nx.draw_networkx_edges(G, pos, width=2.5, alpha=1, edge_color='gray')
    nx.draw_networkx_labels(G, pos, font_size=18, font_weight='bold')

    plt.title("Mappa del Mondo - Continenti")
    plt.axis('off')
    plt.show()

```

```python
from pyswip import Prolog

# --- AGGIORNAMENTO INTERFACCIA PROLOG ---
class RisikoPrologInterface:
    def __init__(self):
        self.prolog = Prolog()
        self._load_knowledge_base()

    def _load_knowledge_base(self):
        try:
            self.prolog.consult("risiko_geografia.pl")
            self.prolog.consult("risiko_obbiettivi.pl")
            self.prolog.consult("risiko_tattica.pl")
            self.prolog.consult("risiko_analisi_globale.pl")
            self.prolog.consult("risiko_comportamento.pl")
            print("KB Completa caricata (Geo, Obiettivi, Tattica, Globale, Comportamento).")
        except Exception as e:
            print(f"Errore caricamento KB: {e}")

    def sync_state(self, risk_game_instance):
        board = risk_game_instance.board if hasattr(risk_game_instance, 'board') else risk_game_instance

        self.prolog.retractall("occupato(_,_,_)")
        self.prolog.retractall("carte_giocatore(_,_)")
        self.prolog.retractall("conquista_effettuata(_)")
        self.prolog.retractall("bonus_carte_attivo(_)")

        for node in board.graph.nodes():
            owner = board.get_owner(node)
            armies = board.get_armies(node)
            if owner is not None and armies is not None:
                self.prolog.assertz(f"occupato({node}, {owner}, {armies})")

        if hasattr(risk_game_instance, 'hands'):
            current = risk_game_instance.current_player
            if risk_game_instance.conquest_this_turn:
                self.prolog.assertz(f"conquista_effettuata({current})")

            for player, cards in risk_game_instance.hands.items():
                num_cards = len(cards)
                self.prolog.assertz(f"carte_giocatore({player}, {num_cards})")
                if risk_game_instance.has_tris_bonus(player):
                    self.prolog.assertz(f"bonus_carte_attivo({player})")

    # === QUERY WRAPPERS ===

    def should_play_tris(self, player): # ### NUOVO ###
        return list(self.prolog.query(f"conviene_giocare_tris({player})"))

    def should_force_attack(self, player):
        return list(self.prolog.query(f"forza_attacco({player})"))

    def query_objective_status(self, player, objective_struct):
        res = list(self.prolog.query(f"quanto_manca({player}, {objective_struct}, X)"))
        return res[0]['X'] if res else None

    def get_vulnerable_territories(self, player):
        res = list(self.prolog.query(f"vulnerabile(T, {player})"))
        return list(set([r['T'] for r in res]))

    def get_frontline_territories(self, player):
        res = list(self.prolog.query(f"fronte(T, {player})"))
        return list(set([r['T'] for r in res]))

    def get_player_state(self, player):
        states = []
        if list(self.prolog.query(f"dominante({player})")): states.append("DOMINANTE")
        if list(self.prolog.query(f"in_svantaggio({player})")): states.append("IN_SVANTAGGIO")
        if list(self.prolog.query(f"normale({player})")): states.append("NORMALE")
        if list(self.prolog.query(f"rischio_eliminazione({player})")): states.append("RISCHIO_ELIMINAZIONE")
        return states

    def get_player_score(self, player):
        res = list(self.prolog.query(f"calcola_punteggio({player}, X)"))
        return res[0]['X'] if res else 0.0

    def get_contested_continents(self):
        query = "continente_conteso(C)"
        result = list(self.prolog.query(query))
        return [res['C'] for res in result]

    def get_fortress_continents(self):
        query = "continente_fortezza(C, G)"
        result = list(self.prolog.query(query))
        return {res['C']: res['G'] for res in result}

    def get_strategy_mode(self, player):
        res = list(self.prolog.query(f"debug_modalita({player}, M)"))
        return res[0]['M'] if res else "sconosciuta"

    def get_action_priority(self, player, action_type):
        query = f"ottieni_priorita({player}, {action_type}, Peso)"
        res = list(self.prolog.query(query))
        return res[0]['Peso'] if res else 0

    def get_strategic_entry_points(self, player, objective_struct):
        targets = set()
        query_continents = (
            f"({objective_struct} = conquista_continenti(C1, C2)), "
            f"intercontinentale(T), "
            f"(territorio(T, C1) ; territorio(T, C2)), "
            f"\\+ occupato(T, {player}, _)"
        )
        try:
            res = list(self.prolog.query(query_continents))
            for r in res: targets.add(r['T'])
        except Exception: pass

        query_destroy = (
            f"({objective_struct} = distruggi_giocatore(Nemico)), "
            f"occupato(T, Nemico, _), "
            f"adiacente(T, Mio), "
            f"occupato(Mio, {player}, _)"
        )
        try:
            res = list(self.prolog.query(query_destroy))
            for r in res: targets.add(r['T'])
        except Exception: pass
        return list(targets)

    def get_objective_scope(self, player, objective_struct):
        query = f"territorio_obiettivo({player}, {objective_struct}, T)"
        result = list(self.prolog.query(query))
        return list(set([res['T'] for res in result]))

    def debug_knowledge(self, players):
        report = {}
        for p in players:
            mode = self.get_strategy_mode(p)
            vuln = len(self.get_vulnerable_territories(p))
            front = len(self.get_frontline_territories(p))
            states = self.get_player_state(p)
            score = self.get_player_score(p)
            report[p] = {
                "Modalità": mode,
                "Vulnerabili": vuln,
                "Fronte": front,
                "Tags": states,
                "PowerScore": score
            }
        return report
```

```python
import itertools

# =============================================================================
# STRUMENTO 1: CALCOLATORE PROBABILISTICO (MARKOV + RISK ITALIANO)
# =============================================================================

class RiskProbabilityCalculator:
    def __init__(self):
        self.bmem = {} # Battle memoization
        self.omem = {} # Outcome memoization
        self.tmem = {} # Tour memoization
        # Pre-calcola la matrice delle probabilità dei dadi all'avvio
        self.probs = self._precompute_dice_probabilities()

    def _precompute_dice_probabilities(self):
        """
        Genera la matrice delle probabilità per scontri fino a 3v3.
        Regole Italiane: Difensore vince i pareggi, max 3 dadi per parte.
        """
        # probs[att_dadi][def_dadi] = lista probabilità perdite attaccante [0, 1, 2, 3]
        probs = [[None for _ in range(4)] for _ in range(4)]
        dice_sides = [1, 2, 3, 4, 5, 6]

        for n_att in range(1, 4):
            for n_def in range(1, 4):
                att_rolls = list(itertools.product(dice_sides, repeat=n_att))
                def_rolls = list(itertools.product(dice_sides, repeat=n_def))
                total = len(att_rolls) * len(def_rolls)
                attacker_losses_count = [0] * (min(n_att, n_def) + 1)

                for a_roll in att_rolls:
                    for d_roll in def_rolls:
                        a_sorted = sorted(a_roll, reverse=True)
                        d_sorted = sorted(d_roll, reverse=True)
                        losses = 0
                        for k in range(min(n_att, n_def)):
                            # Regola: Difensore vince i pareggi
                            if d_sorted[k] >= a_sorted[k]:
                                losses += 1
                        attacker_losses_count[losses] += 1

                probs[n_att][n_def] = [c / total for c in attacker_losses_count]
        return probs

    def outcome_prob(self, attackers, defenders, arem=1, drem=0):
        """
        Calcola ricorsivamente la probabilità che uno scontro finisca
        esattamente con 'arem' attaccanti e 'drem' difensori rimasti.
        """
        if attackers < arem or defenders < drem: return 0.0
        if defenders == drem: return 1.0 if attackers == arem else 0.0

        h = (attackers, defenders, arem, drem)
        if h in self.omem: return self.omem[h]

        # Determina numero dadi (Attaccante lascia sempre 1 armata di base sul territorio di lancio)
        if attackers >= 4: n_att = 3
        elif attackers == 3: n_att = 2
        elif attackers == 2: n_att = 1
        else: return 0.0 # Non può attaccare

        n_def = min(3, defenders) # Difesa italiana: max 3 dadi

        outcome_probs = self.probs[n_att][n_def]
        val = 0.0
        comparisons = min(n_att, n_def)

        for att_lost, prob in enumerate(outcome_probs):
            if prob > 0:
                def_lost = comparisons - att_lost
                val += prob * self.outcome_prob(attackers - att_lost, defenders - def_lost, arem, drem)

        self.omem[h] = val
        return val

    def tour_prob(self, attackers, darmies, tindex=0, fortify=None):
        """
        Calcola la probabilità di vincere una sequenza di territori (darmies).
        fortify: lista di quante armate lasciare su ogni territorio conquistato.
        """
        if fortify is None: fortify = [1] * len(darmies)

        if tindex == len(darmies): return 1.0 # Tutti i territori conquistati
        if tindex == 0: self.tmem = {} # Reset cache per nuova catena

        h = (attackers, tindex)
        if h in self.tmem: return self.tmem[h]

        # Armate disponibili per questo scontro:
        current_attackers = attackers
        enemy_armies = darmies[tindex]

        # Check rapido: Ho abbastanza truppe per coprire i presidi futuri?
        future_fortify_cost = sum(fortify[i] for i in range(tindex, len(darmies)))
        if current_attackers <= future_fortify_cost: return 0.0

        val = 0.0
        # Somma su tutti i possibili esiti positivi (sopravvissuti > 0)
        # i = armate sopravvissute DOPO aver sconfitto questo nemico
        for i in range(1, current_attackers + 1):
             prob_outcome = self.outcome_prob(current_attackers, enemy_armies, i)
             if prob_outcome > 0:
                 # Passo successivo: Sottraggo il presidio (fortify[tindex]) da lasciare QUI
                 next_attackers = i - fortify[tindex]
                 if next_attackers > 0:
                     val += prob_outcome * self.tour_prob(next_attackers, darmies, tindex+1, fortify)

        self.tmem[h] = val
        return val

    # --- METODO MANCANTE CHE CAUSAVA L'ERRORE ---
    def battle_prob(self, attackers, defenders, min_fortify=1):
        """
        Wrapper per calcolare la probabilità di vittoria di una singola battaglia.
        attackers: numero TOTALE di armate nel territorio attaccante.
        defenders: numero di armate nel territorio difensore.
        """
        # Se ho X armate totali, ne posso usare X-1 per l'attacco effettivo
        if attackers <= 1: return 0.0
        mobile_troops = attackers - 1

        # Chiamiamo tour_prob passando il nemico come una lista di un solo elemento [defenders]
        return self.tour_prob(mobile_troops, [defenders], fortify=[min_fortify])

# --- FUNZIONE HELPER PER INTEGRAZIONE AGENTE ---

def analyze_attack_chain(calculator, board, player, start_node, chain_nodes, available_troops):
    """
    Analizza una catena di attacco [T1, T2...] partendo da start_node.
    Definisce dinamicamente quante truppe lasciare (Fronte vs Retrovia).
    """
    darmies = []
    fortify_plan = []
    step_details = []

    # Simuliamo l'avanzamento per capire la topologia futura
    # Nota: start_node è MIO. chain_nodes sono NEMICI.

    # 1. Analisi Topologica (Costruzione Piano Presidi)
    simulated_owner = {n: player for n in [start_node] + chain_nodes} # Ipotizziamo di averli presi tutti

    full_path = [start_node] + chain_nodes

    for i in range(len(chain_nodes)):
        current = full_path[i]       # Territorio che abbiamo appena preso (o start)
        next_target = full_path[i+1] # Prossimo da attaccare

        # Quante armate ha il nemico?
        enemy_units = board.get_armies(next_target)
        darmies.append(enemy_units)

        # Analisi: Una volta spostato l'attacco su 'next_target', 'current' sarà sicuro?
        # Guardiamo i vicini di 'current'.
        neighbors = board.get_neighbors(current)
        is_frontline = False

        for n in neighbors:
            if n == next_target: continue

            real_owner = board.get_owner(n)
            effective_owner = simulated_owner.get(n, real_owner)

            if effective_owner != player:
                is_frontline = True
                break

        leave_behind = 3 if is_frontline else 1
        fortify_plan.append(leave_behind)

        status = "FRONTE" if is_frontline else "RETROVIA"
        step_details.append({
            'step': f"{current} -> {next_target}",
            'enemies': enemy_units,
            'leave_on_origin': leave_behind,
            'status_origin': status
        })

    # 2. Calcolo Probabilità
    mobile_troops = available_troops - 1

    final_res = {'total_prob': 0.0, 'steps': step_details, 'fortify_plan': fortify_plan}

    # Probabilità cumulativa fino allo step i
    for i in range(1, len(chain_nodes) + 1):
        sub_darmies = darmies[:i]
        sub_fortify = fortify_plan[:i]

        prob = calculator.tour_prob(mobile_troops, sub_darmies, 0, sub_fortify)
        step_details[i-1]['cumul_prob'] = prob

    final_res['total_prob'] = step_details[-1]['cumul_prob']
    return final_res
```

```python
import heapq
from collections import deque

# =============================================================================
# STRUMENTO 2: PATHFINDER STRATEGICO (A* CON VINCOLO INVASIONE)
# =============================================================================

class StrategicPathfinder:
    def __init__(self, risk_board, prob_calculator, kb_interface):
        self.board = risk_board
        self.calculator = prob_calculator
        self.kb = kb_interface

    def heuristic(self, start_node, goal_node):
        """
        Calcola la distanza topologica (numero minimo di territori).
        È un'euristica AMMISSIBILE per A* perché ogni territorio nemico
        ha almeno 1 armata, quindi: Distanza <= Costo Reale (Armate).
        """
        if start_node == goal_node:
            return 0

        # Semplice BFS per trovare la distanza minima in "salti"
        queue = deque([(start_node, 0)])
        visited = {start_node}

        while queue:
            current, dist = queue.popleft()
            if current == goal_node:
                return dist

            for neighbor in self.board.get_neighbors(current):
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append((neighbor, dist + 1))

        return float('inf')

    def a_star_attack_chain(self, start_node, goal_node, player_color):
        """
        Trova la catena di attacco ottimale (minor resistenza nemica)
        VINCOLO: Non si può passare attraverso territori già alleati.
        """
        frontier = []
        heapq.heappush(frontier, (0, start_node))

        came_from = {start_node: None}
        cost_so_far = {start_node: 0}

        while frontier:
            # Nota: current_cost qui è f(n), ma noi usiamo cost_so_far per g(n)
            current_cost, current_node = heapq.heappop(frontier)

            if current_node == goal_node:
                break

            neighbors = self.board.get_neighbors(current_node)
            for next_node in neighbors:
                owner = self.board.get_owner(next_node)

                # --- APPLICAZIONE VINCOLO ---
                if owner == player_color:
                    continue

                # --- CALCOLO COSTO (RESISTENZA) ---
                armies = self.board.get_armies(next_node)
                weight = armies + 1

                new_cost = cost_so_far[current_node] + weight

                if next_node not in cost_so_far or new_cost < cost_so_far[next_node]:
                    cost_so_far[next_node] = new_cost
                    # f(n) = g(n) + h(n)
                    # Qui invochiamo la nuova euristica definita sopra
                    priority = new_cost + self.heuristic(next_node, goal_node)
                    heapq.heappush(frontier, (priority, next_node))
                    came_from[next_node] = current_node

        # Ricostruzione Percorso
        if goal_node not in came_from:
            return None, float('inf')

        path = []
        curr = goal_node
        while curr != start_node:
            path.append(curr)
            curr = came_from[curr]

        path.reverse()
        return path, cost_so_far[goal_node]

    def evaluate_best_attack_chain(self, player, objective_struct, min_chain_prob=0.2):
        # --- NESSUNA MODIFICA A QUESTO METODO BASE, MA AGGIUNTO PARAMETRO ---
        try:
            starts_query = list(self.kb.prolog.query(f"forte_al_fronte(T, {player})"))
            start_nodes = [res['T'] for res in starts_query]
        except:
            start_nodes = []

        if not start_nodes:
            all_my_terrs = self.board.get_territories_by_player(player)
            start_nodes = [t for t in all_my_terrs if self.board.get_armies(t) >= 4]

        objective_scope = self.kb.get_objective_scope(player, objective_struct)
        goal_nodes = [t for t in objective_scope if self.board.get_owner(t) != player]

        if not start_nodes or not goal_nodes:
            return None

        best_chain_report = None
        best_prob = -1.0

        print(f"   [Pathfinder] Analisi: {len(start_nodes)} Sorgenti x {len(goal_nodes)} Destinazioni...")

        for start in start_nodes:
            available_troops = self.board.get_armies(start)

            for goal in goal_nodes:
                path, resistance = self.a_star_attack_chain(start, goal, player)

                if path:
                    report = analyze_attack_chain(
                        self.calculator,
                        self.board,
                        player,
                        start,
                        path,
                        available_troops
                    )

                    prob = report['total_prob']

                    # +++ MODIFICA: Uso del parametro min_chain_prob +++
                    if prob > best_prob and prob > min_chain_prob:
                        best_prob = prob
                        best_chain_report = {
                            'start': start,
                            'path': path,
                            'resistance_cost': resistance,
                            'probability_report': report
                        }

        return best_chain_report
```

```python
# --- VERSIONE SILENZIOSA DEL PATHFINDER ---
class SilentStrategicPathfinder(StrategicPathfinder):
    """Eredita tutto dal pathfinder originale, ma rimuove le print di analisi."""

    # +++ MODIFICA: Accetta min_chain_prob +++
    def evaluate_best_attack_chain(self, player, objective_struct, min_chain_prob=0.2):
        try:
            starts_query = list(self.kb.prolog.query(f"forte_al_fronte(T, {player})"))
            start_nodes = [res['T'] for res in starts_query]
        except:
            start_nodes = []

        if not start_nodes:
            all_my_terrs = self.board.get_territories_by_player(player)
            start_nodes = [t for t in all_my_terrs if self.board.get_armies(t) >= 4]

        objective_scope = self.kb.get_objective_scope(player, objective_struct)
        goal_nodes = [t for t in objective_scope if self.board.get_owner(t) != player]

        if not start_nodes or not goal_nodes:
            return None

        best_chain_report = None
        best_prob = -1.0

        # RIMOSSO: print(f" [Pathfinder] Analisi...")

        for start in start_nodes:
            available_troops = self.board.get_armies(start)
            for goal in goal_nodes:
                path, resistance = self.a_star_attack_chain(start, goal, player)
                if path:
                    report = analyze_attack_chain(self.calculator, self.board, player, start, path, available_troops)
                    prob = report['total_prob']
                    # +++ MODIFICA: Uso parametro min_chain_prob +++
                    if prob > best_prob and prob > min_chain_prob:
                        best_prob = prob
                        best_chain_report = {
                            'start': start,
                            'path': path,
                            'resistance_cost': resistance,
                            'probability_report': report
                        }
        return best_chain_report

# --- AGENTE LOGICO SILENZIOSO CON IPERPARAMETRI ---
class SilentLogicalRiskAgent:
    def __init__(self, name, color, objective_prolog_struct, game_engine, kb_interface, prob_calculator, hyperparams=None):
        self.name = name
        self.color = color
        self.objective = objective_prolog_struct
        self.game = game_engine
        self.kb = kb_interface
        self.calculator = prob_calculator
        self.pathfinder = SilentStrategicPathfinder(game_engine.board, prob_calculator, kb_interface)

        default_params = {
            'min_battle_win_prob': 0.40,
            'min_chain_win_prob': 0.20,
            'w_strategic_entry': 50,
            'w_objective_scope': 25,
            'w_fortress_penalty': -30
        }
        if hyperparams:
            self.params = {**default_params, **hyperparams}
        else:
            self.params = default_params

    def play_turn(self):
        self.kb.sync_state(self.game)
        self._execute_placement_phase()

        self.kb.sync_state(self.game)
        self._execute_attack_phase()

        self.kb.sync_state(self.game)
        self._execute_maneuver_phase()

        self.game.end_turn()

    def _execute_placement_phase(self):
        # ### NUOVO: LOGICA SCAMBIO TRIS ###
        # Prima di piazzare, controlla se conviene giocare le carte
        if self.kb.should_play_tris(self.color):
            bonus = self.game.redeem_cards(self.color)
            # if bonus > 0: print(f"  [AI] {self.color} ha giocato un tris! (+{bonus} armate)")

        troops = self.game.troops_to_place
        if troops == 0: return

        w_difesa = self.kb.get_action_priority(self.color, "rinforzo_difensivo")
        w_attacco = self.kb.get_action_priority(self.color, "rinforzo_offensivo")
        w_totale = w_difesa + w_attacco
        if w_totale == 0: w_difesa, w_attacco, w_totale = 50, 50, 100

        strategic_targets = self.kb.get_strategic_entry_points(self.color, self.objective)
        objective_scope = self.kb.get_objective_scope(self.color, self.objective)

        quota_difesa = int((troops * w_difesa) / w_totale)
        quota_attacco = troops - quota_difesa

        if quota_difesa > 0:
            vulnerabili = self.kb.get_vulnerable_territories(self.color)
            if not vulnerabili:
                quota_attacco += quota_difesa
            else:
                vuln_importanti = [v for v in vulnerabili if v in objective_scope]
                candidati = vuln_importanti if vuln_importanti else vulnerabili
                while quota_difesa > 0:
                    self.game.place_troops(random.choice(candidati), 1)
                    quota_difesa -= 1

        if quota_attacco > 0:
            frontline = self.kb.get_frontline_territories(self.color)
            if not frontline: frontline = self.game.board.get_territories_by_player(self.color)
            strategic_frontline = []
            for t in frontline:
                neighbors = self.game.board.get_neighbors(t)
                is_gateway = any(n in strategic_targets for n in neighbors)
                is_inside = t in objective_scope
                if is_gateway or is_inside:
                    strategic_frontline.append(t)
            candidates = strategic_frontline if strategic_frontline else frontline
            while quota_attacco > 0:
                self.game.place_troops(random.choice(candidates), 1)
                quota_attacco -= 1

    def _execute_attack_phase(self):
        w_sicuro = self.kb.get_action_priority(self.color, "attacco_sicuro")
        if w_sicuro < 10: return

        strategic_targets = self.kb.get_strategic_entry_points(self.color, self.objective)
        objective_scope = self.kb.get_objective_scope(self.color, self.objective)
        fortezze = self.kb.get_fortress_continents()
        attacks_limit = 20

        while attacks_limit > 0:
            self.kb.sync_state(self.game)
            best_action = None
            best_score_local = -100

            genius_plan = self.pathfinder.evaluate_best_attack_chain(
                self.color,
                self.objective,
                min_chain_prob=self.params['min_chain_win_prob']
            )

            if genius_plan:
                g_start = genius_plan['start']
                g_next = genius_plan['path'][0]
                if self.game.board.get_owner(g_start) == self.color and self.game.board.get_armies(g_start) > 1:
                    best_action = (g_start, g_next)

            if not best_action:
                candidates = []
                my_terrs = self.game.board.get_territories_by_player(self.color)
                for t in my_terrs:
                    if self.game.board.get_armies(t) > 1:
                        for n in self.game.board.get_neighbors(t):
                            if self.game.board.get_owner(n) != self.color:
                                candidates.append((t, n))
                for src, dst in candidates:
                    att = self.game.board.get_armies(src)
                    defn = self.game.board.get_armies(dst)
                    win_prob = self.calculator.battle_prob(att, defn)

                    if win_prob < self.params['min_battle_win_prob']: continue

                    score_base = int(win_prob * 100)
                    score_obj = self.params['w_strategic_entry'] if dst in strategic_targets else (
                        self.params['w_objective_scope'] if dst in objective_scope else 0
                    )
                    dst_cont = self.game.board.get_territory(dst)['continent']
                    score_fort = self.params['w_fortress_penalty'] if (dst_cont in fortezze and fortezze[dst_cont] != self.color) else 0

                    total_score = score_base + score_obj + score_fort
                    if total_score > best_score_local and total_score > 30:
                        best_score_local = total_score
                        best_action = (src, dst)

            if best_action:
                src, dst = best_action
                conquest = self.game.resolve_battle(src, dst, verbose=False)
                attacks_limit -= 1
            else:
                if self.kb.should_force_attack(self.color):
                    opportunity_action = self._find_opportunity_attack()
                    if opportunity_action:
                        self.game.resolve_battle(opportunity_action[0], opportunity_action[1], verbose=False)
                        break
                break

        self.game.end_attack_phase()

    def _find_opportunity_attack(self):
        candidates = []
        my_terrs = self.game.board.get_territories_by_player(self.color)

        for t in my_terrs:
            if self.game.board.get_armies(t) > 1:
                neighbors = self.game.board.get_neighbors(t)
                for n in neighbors:
                    if self.game.board.get_owner(n) != self.color:
                        enemy_armies = self.game.board.get_armies(n)
                        candidates.append((t, n, enemy_armies))

        candidates.sort(key=lambda x: x[2])

        for src, dst, cost in candidates:
            att = self.game.board.get_armies(src)
            prob = self.calculator.battle_prob(att, cost)
            if prob > 0.60:
                return (src, dst)
        return None

    def _execute_maneuver_phase(self):
        if self.kb.get_action_priority(self.color, "spostamento_fronte") > 0:
            retrovie = [r['T'] for r in self.kb.prolog.query(f"retrovia(T, {self.color})")]
            fronti = self.kb.get_frontline_territories(self.color)
            if retrovie and fronti:
                s = max(retrovie, key=lambda x: self.game.board.get_armies(x))
                t = min(fronti, key=lambda x: self.game.board.get_armies(x))
                amt = self.game.board.get_armies(s) - 1
                if amt > 0 and self.game.board.check_path(s, t, self.color):
                    self.game.fortify(s, t, amt)
```

```python
import numpy as np
import copy
from collections import defaultdict

# =============================================================================
# AGENTI DI CONFRONTO (BASELINE)
# =============================================================================

class ComparativeRandomAgent:
    """
    Agente di base per il confronto.
    Può essere 'random' (puramente casuale) o 'hunter' (focalizzato su un colore target).
    """
    def __init__(self, color, game_engine, strategy='random', target_color=None):
        self.color = color
        self.game = game_engine
        self.strategy = strategy
        self.target_color = target_color
        self.name = f"Bot_{color}_{strategy}"

    def play_turn(self):
        # 1. PIAZZAMENTO
        if self.game.troops_to_place > 0:
            my_terrs = self.game.board.get_territories_by_player(self.color)
            if not my_terrs: return

            chosen_territory = None

            # Strategia Hunter: Rinforza i confini col nemico target
            if self.strategy == 'hunter' and self.target_color:
                border_terrs = [t for t in my_terrs if self._has_neighbor_of_color(t, self.target_color)]
                if border_terrs: chosen_territory = np.random.choice(border_terrs)

            # Fallback Random
            if chosen_territory is None: chosen_territory = np.random.choice(my_terrs)

            # Piazza tutto in una volta per semplicità nei bot random
            self.game.place_troops(chosen_territory, self.game.troops_to_place)

        # 2. ATTACCO
        attempts = 20
        while attempts > 0 and self.game.turn_phase == 'attack':
            my_terrs = self.game.board.get_territories_by_player(self.color)
            attack_candidates = []
            priority_candidates = []

            for t in my_terrs:
                if self.game.board.get_armies(t) > 1:
                    neighbors = self.game.board.get_neighbors(t)
                    for n in neighbors:
                        owner_n = self.game.board.get_owner(n)
                        if owner_n != self.color:
                            move = (t, n)
                            attack_candidates.append(move)
                            if self.strategy == 'hunter' and owner_n == self.target_color:
                                priority_candidates.append(move)

            if not attack_candidates: break

            src, dst = None, None
            if priority_candidates:
                idx = np.random.randint(len(priority_candidates))
                src, dst = priority_candidates[idx]
            else:
                idx = np.random.randint(len(attack_candidates))
                src, dst = attack_candidates[idx]

            try:
                # Tenta l'attacco
                won_territory = self.game.resolve_battle(src, dst, verbose=False)
                if not won_territory: attempts -= 1
            except Exception:
                attempts -= 1

        self.game.end_attack_phase()

        # 3. SPOSTAMENTO (Randomico semplice)
        if self.game.turn_phase == 'maneuver':
            my_terrs = self.game.board.get_territories_by_player(self.color)
            # Cerca un territorio con > 1 armate
            movable = [t for t in my_terrs if self.game.board.get_armies(t) > 1]
            if movable:
                src = np.random.choice(movable)
                # Cerca vicini amici
                neighbors = self.game.board.get_neighbors(src)
                friends = [n for n in neighbors if self.game.board.get_owner(n) == self.color]
                if friends:
                    dst = np.random.choice(friends)
                    amt = self.game.board.get_armies(src) - 1
                    if amt > 0:
                        self.game.fortify(src, dst, amt)

        # La chiamata a end_turn è gestita dal ciclo principale o qui,
        # ma per sicurezza i bot chiamano end_turn() implicitamente finendo le fasi.
        self.game.end_turn()

    def _has_neighbor_of_color(self, territory, color):
        for n in self.game.board.get_neighbors(territory):
            if self.game.board.get_owner(n) == color: return True
        return False
```

```python
import numpy as np
import copy
import csv
import os
import matplotlib.pyplot as plt
from collections import defaultdict

# =============================================================================
# AGENTI DI CONFRONTO (BASELINE) - Invariati
# =============================================================================

class ComparativeRandomAgent:
    def __init__(self, color, game_engine, strategy='random', target_color=None):
        self.color = color
        self.game = game_engine
        self.strategy = strategy
        self.target_color = target_color
        self.name = f"Bot_{color}_{strategy}"

    def play_turn(self):
        # 1. PIAZZAMENTO
        if self.game.troops_to_place > 0:
            my_terrs = self.game.board.get_territories_by_player(self.color)
            if not my_terrs: return
            chosen_territory = None
            if self.strategy == 'hunter' and self.target_color:
                border_terrs = [t for t in my_terrs if self._has_neighbor_of_color(t, self.target_color)]
                if border_terrs: chosen_territory = np.random.choice(border_terrs)
            if chosen_territory is None: chosen_territory = np.random.choice(my_terrs)
            self.game.place_troops(chosen_territory, self.game.troops_to_place)

        # 2. ATTACCO
        attempts = 20
        while attempts > 0 and self.game.turn_phase == 'attack':
            my_terrs = self.game.board.get_territories_by_player(self.color)
            attack_candidates = []
            priority_candidates = []
            for t in my_terrs:
                if self.game.board.get_armies(t) > 1:
                    neighbors = self.game.board.get_neighbors(t)
                    for n in neighbors:
                        owner_n = self.game.board.get_owner(n)
                        if owner_n != self.color:
                            move = (t, n)
                            attack_candidates.append(move)
                            if self.strategy == 'hunter' and owner_n == self.target_color:
                                priority_candidates.append(move)
            if not attack_candidates: break
            src, dst = None, None
            if priority_candidates:
                idx = np.random.randint(len(priority_candidates))
                src, dst = priority_candidates[idx]
            else:
                idx = np.random.randint(len(attack_candidates))
                src, dst = attack_candidates[idx]
            try:
                won_territory = self.game.resolve_battle(src, dst, verbose=False)
                if not won_territory: attempts -= 1
            except Exception: attempts -= 1
        self.game.end_attack_phase()

        # 3. SPOSTAMENTO
        if self.game.turn_phase == 'maneuver':
            my_terrs = self.game.board.get_territories_by_player(self.color)
            movable = [t for t in my_terrs if self.game.board.get_armies(t) > 1]
            if movable:
                src = np.random.choice(movable)
                neighbors = self.game.board.get_neighbors(src)
                friends = [n for n in neighbors if self.game.board.get_owner(n) == self.color]
                if friends:
                    dst = np.random.choice(friends)
                    amt = self.game.board.get_armies(src) - 1
                    if amt > 0: self.game.fortify(src, dst, amt)
        self.game.end_turn()

    def _has_neighbor_of_color(self, territory, color):
        for n in self.game.board.get_neighbors(territory):
            if self.game.board.get_owner(n) == color: return True
        return False

# =============================================================================
# RISK ARENA: REPORTISTICA SCIENTIFICA
# =============================================================================

class RiskArena:
    # ### MODIFICA: Aggiunto parametro output_dir ###
    def __init__(self, kb_interface, prob_calculator, output_dir='.'):
        self.kb = kb_interface
        self.calc = prob_calculator
        self.game_logs = []
        self.output_dir = '/content/drive/My Drive/Salvati/Risiko_Simulations' # ### MODIFICA: Salviamo il percorso ###

        # ### MODIFICA: Assicuriamo che la cartella esista ###
        if not os.path.exists(self.output_dir):
            os.makedirs(self.output_dir)
            print(f"[INIT] Creata cartella di output: {self.output_dir}")
        else:
            print(f"[INIT] Output impostato su: {self.output_dir}")


    def _sync_kb(self, game):
        self.kb.sync_state(game)

    def _count_continents(self, game, player):
        count = 0
        for cont_name in game.board.continents:
            if game.board.is_continent_owned(cont_name, player):
                count += 1
        return count

    def _get_kb_metric_count(self, player, metric_predicate):
        query = f"{metric_predicate}(T, {player})"
        try:
            res = list(self.kb.prolog.query(query))
            return len(set([r['T'] for r in res]))
        except Exception:
            return 0

    def run_simulation(self, scenario_name, blue_objective_prolog, n_games=50, agent_config=None):
        print(f"\n{'='*60}")
        print(f" AVVIO SIMULAZIONE AVANZATA: {scenario_name}")
        print(f" Target: {n_games} partite.")
        print(f"{'='*60}")

        self.game_logs = []

        for i in range(n_games):
            game = RiskGame()

            agent_blue = SilentLogicalRiskAgent("AI_Blu", "blu", blue_objective_prolog, game, self.kb, self.calc, hyperparams=agent_config)
            agent_red = ComparativeRandomAgent("rosso", game, strategy='random')
            agent_green = ComparativeRandomAgent("verde", game, strategy='random')
            agents_map = {'blu': agent_blue, 'rosso': agent_red, 'verde': agent_green}
            players_list = ['blu', 'rosso', 'verde']

            state_tracker = {p: {'dominante': 0, 'rischio': 0} for p in players_list}

            winner = None
            turn_count = 0
            max_turns = 200

            while turn_count < max_turns and winner is None:
                current_color = game.current_player
                active_agent = agents_map[current_color]

                if len(game.board.get_territories_by_player(current_color)) == 0:
                    game.end_turn()
                    if game.current_player_idx == 0: turn_count += 1
                    continue

                try:
                    active_agent.play_turn()
                except Exception:
                    game.end_turn()

                self._sync_kb(game)
                for p in players_list:
                    states = self.kb.get_player_state(p)
                    if 'DOMINANTE' in states:
                        state_tracker[p]['dominante'] += 1
                    if 'RISCHIO_ELIMINAZIONE' in states:
                        state_tracker[p]['rischio'] += 1

                if current_color == 'blu':
                    if list(self.kb.prolog.query(f"obiettivo_raggiunto(blu, {blue_objective_prolog})")):
                        winner = 'blu'

                active = [p for p in players_list if len(game.board.get_territories_by_player(p)) > 0]
                if len(active) == 1:
                    winner = active[0]

                if game.current_player_idx == 0: turn_count += 1

            if winner is None: winner = "draw"

            self._sync_kb(game)

            log_entry = {
                'game_id': i + 1,
                'winner': winner,
                'turns': turn_count,
                'objective_blue': blue_objective_prolog,
            }

            for p in players_list:
                terrs = game.board.get_territories_by_player(p)
                log_entry[f'{p}_num_territori'] = len(terrs)
                log_entry[f'{p}_lista_territori'] = str(terrs)
                log_entry[f'{p}_num_armate'] = sum([game.board.get_armies(t) for t in terrs])
                log_entry[f'{p}_num_carte'] = len(game.hands[p])
                log_entry[f'{p}_continenti'] = self._count_continents(game, p)
                log_entry[f'{p}_retrovia'] = self._get_kb_metric_count(p, "retrovia")
                log_entry[f'{p}_forte_fronte'] = self._get_kb_metric_count(p, "forte_al_fronte")
                log_entry[f'{p}_times_dominante'] = state_tracker[p]['dominante']
                log_entry[f'{p}_times_rischio'] = state_tracker[p]['rischio']

            if agent_config:
                for k, v in agent_config.items():
                    log_entry[f'hp_{k}'] = v
            else:
                 default_hp = ['min_battle_win_prob', 'min_chain_win_prob', 'w_strategic_entry', 'w_objective_scope', 'w_fortress_penalty']
                 for k in default_hp: log_entry[f'hp_{k}'] = getattr(agent_blue, 'params', {}).get(k, '')

            self.game_logs.append(log_entry)

            # Stampa ogni partita per capire la velocità
            print(f" > Partita {i+1}/{n_games} completata ({winner})")

        self.export_data(scenario_name, agent_config)

    def export_data(self, scenario_name, agent_config):
        base_name = f"risiko_stats_{scenario_name.replace(' ', '_')}"

        # ### MODIFICA: Uso os.path.join con self.output_dir per tutti i file ###
        csv_file = os.path.join(self.output_dir, f"{base_name}.csv")
        txt_file = os.path.join(self.output_dir, f"{base_name}_summary.txt")
        plot_file = os.path.join(self.output_dir, f"{base_name}_plots.png")

        # 1. SALVATAGGIO CSV
        if not self.game_logs: return
        keys = self.game_logs[0].keys()

        with open(csv_file, 'w', newline='') as f:
            dict_writer = csv.DictWriter(f, keys)
            dict_writer.writeheader()
            dict_writer.writerows(self.game_logs)

        print(f"\n[EXPORT] Dati grezzi salvati in: {csv_file}")

        # 2. CALCOLO STATISTICHE
        wins = defaultdict(int)
        turns = []
        terr_final = defaultdict(list)

        for log in self.game_logs:
            wins[log['winner']] += 1
            if log['winner'] != 'draw':
                turns.append(log['turns'])
            for p in ['blu', 'rosso', 'verde']:
                terr_final[p].append(log[f'{p}_num_territori'])

        n_games = len(self.game_logs)

        # 3. GENERAZIONE REPORT TXT
        with open(txt_file, 'w') as f:
            f.write(f"REPORT SIMULAZIONE: {scenario_name}\n")
            f.write(f"Partite: {n_games}\n")
            f.write(f"Configurazione AI: {agent_config}\n")
            f.write("-" * 40 + "\n")

            f.write("RISULTATI VITTORIA:\n")
            for p in ['blu', 'rosso', 'verde', 'draw']:
                rate = (wins[p] / n_games) * 100
                f.write(f"  {p.upper():<10}: {wins[p]:3d} ({rate:5.1f}%)\n")

            f.write("\nSTATISTICHE TURNI (Solo partite vinte):\n")
            if turns:
                f.write(f"  Media: {np.mean(turns):.2f}\n")
                f.write(f"  StdDev: {np.std(turns):.2f}\n")
                f.write(f"  Min/Max: {np.min(turns)} / {np.max(turns)}\n")

            f.write("\nTERRITORI FINALI (Media +/- StdDev):\n")
            for p in ['blu', 'rosso', 'verde']:
                data = terr_final[p]
                f.write(f"  {p.upper():<10}: {np.mean(data):5.2f} +/- {np.std(data):5.2f}\n")

        print(f"[EXPORT] Report statistico salvato in: {txt_file}")

        # 4. GENERAZIONE GRAFICI
        self._plot_statistics(plot_file, wins, terr_final, turns) # ### MODIFICA: passo il path completo ###

    def _plot_statistics(self, plot_file, wins, terr_final, turns):
        # A. Pie Chart Vittorie
        plt.figure(figsize=(10, 5))

        plt.subplot(1, 2, 1)
        labels = [k for k in wins.keys() if wins[k] > 0]
        sizes = [wins[k] for k in labels]
        colors_map = {'blu': '#4169E1', 'rosso': '#DC143C', 'verde': '#228B22', 'draw': 'grey'}
        colors = [colors_map.get(l, 'grey') for l in labels]

        plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
        plt.title('Distribuzione Vittorie')

        # B. Box Plot Territori
        plt.subplot(1, 2, 2)
        data_to_plot = [terr_final['blu'], terr_final['rosso'], terr_final['verde']]
        plt.boxplot(data_to_plot, labels=['Blu (AI)', 'Rosso', 'Verde'], patch_artist=True)
        plt.title('Distribuzione Territori Finali')
        plt.grid(True, linestyle='--', alpha=0.7)

        plt.tight_layout()
        plt.savefig(plot_file) # ### MODIFICA: Usa il path completo ###
        plt.close()
        print(f"[EXPORT] Grafici salvati in: {plot_file}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
import glob
import os
from google.colab import drive

# --- MODIFICA: Assicuriamo che Drive sia montato ---
drive.mount('/content/drive')

# =============================================================================
# ANALISI BATCH: AGGREGAZIONE E FEATURE IMPORTANCE PER SCENARIO
# =============================================================================

def analyze_aggregated_scenarios():
    # --- MODIFICA: Definizione della cartella target ---
    target_dir = '/content/drive/My Drive/Salvati/Risiko_Simulations'
    #target_dir = '/content/simulazioni caricate'

    # Definiamo gli ID degli scenari attesi (come da Grid Search)
    scenario_ids = [
        "SCENARIO_1", "SCENARIO_2", "SCENARIO_3", "SCENARIO_4",
        "SCENARIO_5", "SCENARIO_6", "SCENARIO_7", "SCENARIO_8"
    ]

    print(f"🔄 Avvio aggregazione CSV e analisi su {len(scenario_ids)} Scenari...")
    print(f"📂 Cartella di ricerca: {target_dir}\n")

    for scen_id in scenario_ids:
        # 1. COSTRUZIONE PATTERN E RICERCA FILE
        # --- MODIFICA: Il pattern ora include il percorso completo della cartella target ---
        # Cerca: /content/drive/.../risiko_stats_SCENARIO_X_*.csv
        pattern = os.path.join(target_dir, f"risiko_stats_{scen_id}_*.csv")

        files = glob.glob(pattern)

        # Filtro di sicurezza: escludiamo eventuali file che non c'entrano
        # Assicuriamoci che siano file della grid search (contengono parentesi quadre per il profilo)
        valid_files = [f for f in files if '[' in f and ']' in f]

        if not valid_files:
            print(f"⚠️ Nessun file valido trovato per {scen_id} in {target_dir}. Skippo.")
            continue

        print(f"{'='*80}")
        print(f"📊 ANALISI {scen_id} (Dataset unificato da {len(valid_files)} profili)")
        # Mostriamo solo il nome file nel log per pulizia, non tutto il percorso
        print(f"   File sorgente: {[os.path.basename(f) for f in valid_files]}")
        print(f"{'='*80}")

        # 2. UNIONE (CONCATENAZIONE) DEI DATAFRAME
        dfs = []
        for f in valid_files:
            try:
                df_temp = pd.read_csv(f)

                # Estraiamo il nome del profilo dal nome file per tracciabilità
                # Es: risiko_stats_SCENARIO_1_[PROFILE_AGGRESSIVO].csv -> PROFILE_AGGRESSIVO
                filename = os.path.basename(f)
                if '[' in filename and ']' in filename:
                    profile_name = filename.split('[')[1].split(']')[0]
                else:
                    profile_name = "unknown"

                df_temp['source_profile'] = profile_name
                dfs.append(df_temp)
            except Exception as e:
                print(f"   ❌ Errore lettura {f}: {e}")

        if not dfs: continue

        # Creazione del Dataset Unificato per questo Scenario
        full_df = pd.concat(dfs, ignore_index=True)
        total_games = len(full_df)
        print(f"   Totale Partite nel Dataset: {total_games}")

        # 3. PREPROCESSING PER MACHINE LEARNING
        # Target: 1 se ha vinto 'blu', 0 altrimenti
        if 'winner' not in full_df.columns:
            print("   [Errore] Colonna 'winner' mancante nel CSV.")
            continue

        full_df['target_win'] = (full_df['winner'] == 'blu').astype(int)

        # Check Bilanciamento Classi
        win_rate = full_df['target_win'].mean()
        if win_rate == 0.0:
            print(f"   ⚠️ DATASET SBILANCIATO: Il Blu non ha MAI vinto in {total_games} partite.")
            print("   Impossibile addestrare il modello (manca la classe positiva).")
            continue
        if win_rate == 1.0:
            print(f"   ⚠️ DATASET SBILANCIATO: Il Blu ha SEMPRE vinto.")
            print("   Impossibile distinguere i parametri migliori (tutti vincono).")
            continue

        # Selezione Features: Tutte le colonne che iniziano con 'hp_'
        feature_cols = [c for c in full_df.columns if c.startswith('hp_')]
        if not feature_cols:
            print("   [Errore] Nessun iperparametro (colonne hp_*) trovato.")
            continue

        X = full_df[feature_cols]
        y = full_df['target_win']

        # 4. RANDOM FOREST (BAGGING) - Feature Importance
        # Usiamo un numero sufficiente di alberi (100) per avere stime stabili
        model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
        model.fit(X, y)

        # 5. ESTRAZIONE E VISUALIZZAZIONE RISULTATI
        importances = pd.DataFrame({
            'Parametro': [col.replace('hp_', '') for col in feature_cols],
            'Importance': model.feature_importances_
        }).sort_values(by='Importance', ascending=False)

        print(f"   Win Rate Globale (Blu): {win_rate*100:.1f}%")
        print("   Fattori decisivi per la vittoria (Top 3):")
        for i, row in importances.head(3).iterrows():
            print(f"   -> {row['Parametro']:<20}: {row['Importance']:.4f}")

        # Grafico
        plt.figure(figsize=(8, 4))
        sns.barplot(x='Importance', y='Parametro', data=importances, palette='viridis', hue='Parametro', legend=False)
        plt.title(f'Analisi Iperparametri: {scen_id}\n(Quali parametri influenzano di più la vittoria?)')
        plt.xlabel('Importanza Relativa (0-1)')
        plt.tight_layout()
        plt.show()
        print("\n")

# Esegui l'analisi aggregata
analyze_aggregated_scenarios()
```

```python
# =============================================================================
# CELLA 1: OTTIMIZZAZIONE PARAMETRI VIA RANDOM FOREST (PURE WIN/LOSS LOGIC)
# =============================================================================
import pandas as pd
import glob
import os
import numpy as np
from sklearn.ensemble import RandomForestRegressor

# --- CONFIGURAZIONE ---
TARGET_SCENARIO = "SCENARIO_1"
DATA_FOLDER = '/content/drive/My Drive/Salvati/Risiko_Simulations'
N_SYNTHETIC_SAMPLES = 100000  # Numero di combinazioni teoriche da testare col modello

print(f"🤖 AVVIO ANALISI: Training Regressore su pure Vittorie/Sconfitte...")

# 1. Caricamento e Unione di TUTTI i Dataset (Massive + Grid)
pattern_std = os.path.join(DATA_FOLDER, f"risiko_stats_{TARGET_SCENARIO}_*.csv")
pattern_val = os.path.join(DATA_FOLDER, f"risiko_stats_VALIDATION_{TARGET_SCENARIO}_*.csv")

all_files = glob.glob(pattern_std) + glob.glob(pattern_val)
# Filtriamo per prendere sia file di Grid Search ('[') sia Validazioni
valid_files = [f for f in list(set(all_files)) if ('[' in f) or ('VALIDATION' in f)]

if not valid_files:
    print(f"❌ NESSUN DATO TROVATO in {DATA_FOLDER}.")
else:
    df_list = []
    for f in valid_files:
        try:
            df_list.append(pd.read_csv(f))
        except Exception: pass

    if df_list:
        full_df = pd.concat(df_list, ignore_index=True)

        # 2. Preparazione Dati per il Regressore
        # INPUT (X): Solo gli iperparametri
        hp_cols = [c for c in full_df.columns if c.startswith('hp_')]

        # TARGET (y): 1.0 se ha vinto 'blu', 0.0 se ha perso.
        # Il Regressore cercherà di minimizzare l'errore verso questi estremi.
        full_df['outcome'] = np.where(full_df['winner'] == 'blu', 1.0, 0.0)

        X = full_df[hp_cols]
        y = full_df['outcome']

        print(f"📊 Dataset: {len(full_df)} partite totali.")
        print(f"🎯 Target: {y.sum()} Vittorie (1.0) vs {len(y) - y.sum()} Sconfitte (0.0).")

        if len(hp_cols) == 0:
            print("❌ Errore: Nessuna colonna iperparametro (hp_) trovata.")
        else:
            # 3. Addestramento Random Forest Regressor
            # Il modello impara la relazione matematica tra Parametri -> Probabilità di Vittoria
            print("🧠 Addestramento Random Forest Regressor in corso...")
            rf_model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
            rf_model.fit(X, y)

            # 4. Stima del Profilo Ottimale (In-Silico Optimization)
            # Generiamo una griglia massiva di combinazioni casuali basate sui range esistenti
            print(f"🔍 Ricerca della configurazione perfetta su {N_SYNTHETIC_SAMPLES} simulazioni virtuali...")

            synthetic_data = {}
            for col in hp_cols:
                min_val = int(full_df[col].min())
                max_val = int(full_df[col].max())
                # Generiamo valori casuali nel range osservato
                synthetic_data[col] = np.random.randint(min_val, max_val + 1, size=N_SYNTHETIC_SAMPLES)

            X_synthetic = pd.DataFrame(synthetic_data)

            # Chiediamo al Regressore: "Quanto è probabile che questa configurazione vinca?"
            # Il valore predetto sarà un float tra 0.0 e 1.0
            predictions = rf_model.predict(X_synthetic)

            # 5. Selezione del Vincitore
            # Prendiamo la riga con la predizione più alta (più vicina a 1.0)
            best_idx = np.argmax(predictions)
            best_config = X_synthetic.iloc[best_idx]
            best_win_prob = predictions[best_idx]

            best_profile = {}

            print("\n🏆 PROFILO OTTIMIZZATO (Stimato dal Regressore):")
            print(f"⭐ Probabilità di Vittoria Stimata: {best_win_prob*100:.2f}%")
            print("-" * 50)
            print(f"{'Parametro':<35} | {'Valore Ottimale'}")
            print("-" * 50)

            for param, val in best_config.items():
                clean_param = param.replace('hp_', '')
                best_profile[clean_param] = int(val)
                print(f"{clean_param:<35} | {int(val)}")

            print("-" * 50)
            print("✅ 'best_profile' pronto.")

    else:
        print("❌ Nessun DataFrame caricato.")
```
